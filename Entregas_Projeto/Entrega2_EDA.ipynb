# **Introdução**

O presente projeto desenvolve-se no contexto da Ciência de Dados aplicada à Economia Industrial. De acordo com Fernandes (2026), uma série temporal é um conjunto de observações ordenadas no tempo e coletadas em intervalos regulares. Neste trabalho, o foco recai sobre o Índice de Produção Física Industrial (PIM-PF), um indicador vital para medir o dinamismo econômico de um país.

## 1.1 Problema

A volatilidade da produção industrial brasileira apresenta um desafio para os gestores de infraestrutura e produção. Sem uma análise preditiva, a indústria enfrenta o risco de subutilização de recursos ou gargalos logísticos. O problema selecionado é a incerteza no planejamento produtivo, que muitas vezes leva ao desperdício de insumos e energia, dificultando a construção de uma infraestrutura resiliente e sustentável.

## 1.2 Motivação e Justificativa

A relevância deste tema fundamenta-se nos Objetivos de Desenvolvimento Sustentável (ODS) da Organização das Nações Unidas (ONU). A capacidade de prever a produção permite que a infraestrutura logística e energética do país seja planejada com maior precisão (ODS 9 – Indústria, Inovação e Infraestrutura). Ao identificar padrões sazonais, as indústrias podem ajustar seu estoque e produção, reduzindo desperdícios de matéria-prima (ODS 12 – Consumo e Produção Responsáveis). A solução proposta consiste em um produto analítico baseado em modelos estatísticos de séries temporais que funcionará como ferramenta de auxílio à tomada de decisão para associações industriais e gestores governamentais.

## 1.3 Objetivo Geral

# **Referencial Teórico**

A análise de séries temporais é um campo da estatística voltado ao estudo de dados coletados sequencialmente ao longo do tempo. Segundo Morettin e Toloi (2018), os principais componentes de uma série temporal são: **tendência** (movimento de longo prazo), **sazonalidade** (padrões que se repetem em períodos fixos) e **ruído** (variações aleatórias não explicadas pelos demais componentes).

Para verificar se uma série é **estacionária** — condição exigida por grande parte dos modelos preditivos —, utiliza-se o **Teste de Dickey-Fuller Aumentado (ADF)**. Uma série é considerada estacionária quando sua média e variância são constantes ao longo do tempo (p-valor ≤ 0,05). Caso contrário, transformações como diferenciação são aplicadas.

Entre as técnicas de modelagem para séries temporais, destacam-se: (1) **modelos ARIMA** (AutoRegressive Integrated Moving Average), amplamente utilizados para séries univariadas estacionárias; (2) **Suavização Exponencial (ETS)**, que pondera observações recentes com maior peso; e (3) **modelos de decomposição clássica**, que separam os componentes da série para análise individual. Cada abordagem apresenta vantagens — ARIMA é flexível para diferentes estruturas de autocorrelação, enquanto ETS é computacionalmente simples e interpretável.

Trabalhos correlatos utilizando o índice PIM-PF do IBGE geralmente empregam ARIMA ou SARIMA (versão sazonal do ARIMA) para capturar tanto a tendência quanto os padrões mensais recorrentes da produção industrial brasileira (MORETTIN; TOLOI, 2018).

# **Pipeline da Solução**

O pipeline de solução é composto pelas seguintes etapas:

1. **Coleta e leitura dos dados**: Importação do arquivo CSV da PIM-PF (SIDRA/IBGE), limpeza do cabeçalho e organização cronológica das colunas.
2. **Pré-processamento**: Conversão de datas, tratamento de valores ausentes e formatação dos índices numéricos.
3. **Análise Exploratória (EDA)**: Visualização da série completa, análise por ano de referência (2012) e identificação visual de padrões.
4. **Decomposição da série temporal**: Separação dos componentes de tendência, sazonalidade e ruído via decomposição aditiva.
5. **Teste de estacionariedade**: Aplicação do Teste ADF para classificar a série.
6. **Modelagem preditiva**: Aplicação de modelo(s) de séries temporais (ARIMA/SARIMA ou ETS) com base nas características identificadas na EDA.
7. **Avaliação e projeção**: Avaliação do modelo por métricas (MAE, RMSE) e geração de previsões futuras.

A seguir, apresenta-se a implementação das etapas 1 a 5 realizadas até esta entrega.

## Etapa 1 e 2 — Coleta, Leitura e Pré-processamento

### Brasil — Série Nacional (PIM_PF_BR_2006_2022)

```python
import pandas as pd
import os

caminho = 'https://raw.githubusercontent.com/MarianaMayumeCaniza/PROJETOAPLICADO4_2026.01/refs/heads/main/BaseDeDados/PIM_PF_BR_2006_2022.csv'

# Pula as primeiras linhas do cabeçalho poluído
df = pd.read_csv(caminho, sep=';', encoding='utf-8', skiprows=3)

# Remove a primeira coluna sem utilidade
df = df.drop(df.columns[0], axis=1)
df = df.iloc[:2]

meses = {
    'janeiro': '01', 'fevereiro': '02', 'março': '03',
    'abril': '04', 'maio': '05', 'junho': '06',
    'julho': '07', 'agosto': '08', 'setembro': '09',
    'outubro': '10', 'novembro': '11', 'dezembro': '12'
}

novas_datas = []
for col in df.columns:
    try:
        mes, ano = col.split()
        mes_num = meses[mes.lower()]
        data = f"{mes_num}/{ano}"
        novas_datas.append(data)
    except:
        novas_datas.append(col)

df.loc['Data'] = novas_datas

# Salva o título da análise
titulo_analise = df.iloc[0, 1]

# Remove linha de título desnecessária e reorganiza
df = df.drop(index=0)
linha_data = df.loc[['Data']]
resto = df.drop(index='Data')
df = pd.concat([linha_data, resto])

print(df)
```

## Etapa 3 — Análise Exploratória: Série Temporal Completa (2006–2022)

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Transposição e conversão
df_final = pd.DataFrame({
    'Data': df.iloc[0].values,
    'indice': df.iloc[1].values
})

df_final = df_final.dropna()
df_final['indice'] = (
    df_final['indice']
    .astype(str)
    .str.replace(',', '.')
    .astype(float)
)

df_final['Data'] = pd.to_datetime(df_final['Data'], format='%m/%Y')
df_final = df_final.set_index('Data')

# Gráfico da série completa
fig, ax = plt.subplots(figsize=(15, 6))
ax.plot(df_final.index, df_final['indice'], label='Índice de Produção', color='tab:blue')
ax.xaxis.set_major_locator(mdates.YearLocator(1))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.xticks(rotation=45)
plt.xlim(df_final.index.min(), df_final.index.max())
plt.title(f'Série Temporal: {titulo_analise}')
plt.ylabel('Número-Índice (2012=100)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Etapa 3 — Análise Exploratória: Série Temporal — Ano Base 2012

```python
df_2012 = df_final.loc['2012']
df_2012.plot(figsize=(12, 5))
plt.title('Série Temporal - 2012 (Ano Base)')
plt.ylabel('Número-Índice (2012=100)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Etapa 4a — Caracterização: Tendência (Média Móvel 12 meses)

```python
# Média móvel de 12 meses para identificar a tendência
df_final['media_movel'] = df_final['indice'].rolling(window=12).mean()

plt.figure(figsize=(12, 5))
plt.plot(df_final.index, df_final['indice'], alpha=0.3, label='Mensal')
plt.plot(df_final.index, df_final['media_movel'], color='red', label='Tendência (12 meses)')
plt.title('Análise de Tendência da Produção Industrial')
plt.ylabel('Número-Índice (2012=100)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Etapa 4b — Caracterização: Sazonalidade (Distribuição por Mês)

```python
import seaborn as sns

df_final['Mes_Num'] = df_final.index.month

plt.figure(figsize=(10, 6))
sns.boxplot(x='Mes_Num', y='indice', data=df_final)
plt.title('Sazonalidade: Distribuição da Produção por Mês (2006–2022)')
plt.xlabel('Mês')
plt.ylabel('Número-Índice (2012=100)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

O boxplot evidencia padrão sazonal consistente: pico de produção nos meses de outubro e novembro, e vale no início do ano (janeiro–fevereiro). O outlier observado no mês 4 (abril) está associado ao impacto da pandemia de COVID-19 em 2020.

## Etapa 5 — Teste de Estacionariedade (Teste ADF)

```python
from statsmodels.tsa.stattools import adfuller

resultado_adf = adfuller(df_final['indice'])
print(f'P-valor do Teste ADF: {resultado_adf[1]:.4f}')

if resultado_adf[1] > 0.05:
    print('Conclusão: A série NÃO é estacionária (p-valor > 0,05). Possui tendência ou variância não constante.')
else:
    print('Conclusão: A série É estacionária (p-valor ≤ 0,05).')
```

## Etapa 4c — Decomposição Clássica (Tendência + Sazonalidade + Ruído)

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Decomposição Aditiva (oscilação sazonal aparentemente constante)
decomposicao = seasonal_decompose(df_final['indice'], model='additive', period=12)

fig = decomposicao.plot()
fig.set_size_inches(12, 8)
plt.suptitle('Decomposição Clássica Aditiva — PIM-PF Brasil', y=1.01)
plt.tight_layout()
plt.show()
```

A decomposição confirma os três componentes identificados visualmente:
- **Tendência**: crescimento até ~2013, queda entre 2014–2016 (crise econômica), recuperação parcial e queda abrupta em 2020 (pandemia).
- **Sazonalidade**: padrão anual regular com amplitude constante, justificando o modelo aditivo.
- **Ruído**: picos de volatilidade em 2008–2009 (crise financeira global) e 2020 (COVID-19).
